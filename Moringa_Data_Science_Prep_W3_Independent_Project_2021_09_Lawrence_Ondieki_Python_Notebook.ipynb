### Call Data Records(CDR) - Analysis

#### CDR for first day from late 2012-05-06 night to earli 2012-05-07 morning

```python
#import data from csv file and make a dataframe.
with open('Telcom_dataset.csv','r') as f:
    ds1_df = pd.read_csv(f, index_col=False, encoding='utf-8')
```

```python
ds1_df
```

```python
# Most used site in late night 2012-05-06 to early morning 2012-05-07
most_used_site=(ds1_df.groupby(['SITE_ID']).sum().loc[lambda df: df['VALUE'] > 0]).sort_values('VALUE',ascending=False)
```

```python
most_df=pd.DataFrame(most_used_site)
```

```python
most_df.head()
```

```python
Cells.set_index('SITE_CODE')
```

```python
city_df=pd.merge(Cells,most_df,left_on='SITE_CODE', right_on='SITE_ID')
```

```python
most_used_city=(city_df.groupby(['VILLES']).sum().loc[lambda df: df['VALUE'] > 0]).sort_values('VALUE',ascending=False)
```

```python
most_used_city.head()
```

```python
ds1_df.loc[(pd.to_datetime(ds1_df['DATETIME']) > pd.Timestamp('2012-05-07 00:00')) & (pd.to_datetime(ds1_df['DATETIME']) < pd.Timestamp('2012-05-07 17:00'))]
```

```python
ds1_df['SITE_ID']
```
